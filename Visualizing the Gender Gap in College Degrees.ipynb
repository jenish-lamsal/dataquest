## In this project, we will visualize the gender gap across college degrees. We will generate line charts to compare gender gap across all degrees. The dataset was scraped from the Department of Education Statistics.

### The code below compares the perecentage of men and women in various subjects.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255) #rgb combination for dark blue
cb_orange = (255/255, 128/255, 14/255) #rgb combination for orange
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']


border = ["right", "left", "top", "bottom"]

fig = plt.figure(figsize=(9, 18))

list_a = [1,4,7,10,13,16]
list_b = [2,5,8,11,14]
list_c = [3,6,9,12,15,18]


for sp in range(1,19):
    if sp in list_a:
        ab = int((sp-1)/3)
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[ab]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[ab]], c=cb_orange, label='Men', linewidth=3)
        for i in border:
            ax.spines[i].set_visible(False) # removes all the borders
        ax.set_xlim(1968, 2011) #setting x-axis range
        ax.set_ylim(0,100) #setting y axis range
        ax.set_title(stem_cats[ab])
        ax.set_yticks([0,100])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off") #removes the ticks
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3) # 50% line
        
        if sp == 1:
            ax.legend(loc = 'best', frameon=False)
        elif sp == 16:
            ax.legend(loc = 'best', frameon=False)
            ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "on")

    elif sp in list_b:
        ab = int((sp-2)/3)
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[ab]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[ab]], c=cb_orange, label='Men', linewidth=3)
        for i in border:
            ax.spines[i].set_visible(False) # removes all the borders
        ax.set_xlim(1968, 2011) #setting x-axis range
        ax.set_ylim(0,100) #setting y axis range
        ax.set_title(lib_arts_cats[ab])
        ax.set_yticks([0,100])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off") #removes the ticks as well as the labels from the x axis
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        
        if sp == 2:
            ax.legend(loc = 'best', frameon=False)
        elif sp == 14:
            ax.legend(loc = 'best', frameon=False)
            ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "on")
            
    elif sp in list_c:
        ab = int((sp-3)/3)
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[ab]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[ab]], c=cb_orange, label='Men', linewidth=3)
        for i in border:
            ax.spines[i].set_visible(False) # removes all the borders
        ax.set_xlim(1968, 2011) #setting x-axis range
        ax.set_ylim(0,100) #setting y axis range
        ax.set_title(other_cats[ab])
        ax.set_yticks([0,100])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off") #removes the ticks
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        
        if sp == 3:
            ax.legend(loc = 'best', frameon=False)
        elif sp == 18:
            ax.legend(loc = 'best', frameon=False)
            ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "on")
            
    else:
        continue
    
        
        
    
        

    
```

### The first column displays gender gap in STEM fields, the second column displays gender gap in liberal arts and the third column displays gender gap in other subjects
